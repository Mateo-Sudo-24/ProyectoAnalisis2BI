```python
import csv
import json
import os

# Lista de archivos CSV a convertir
csv_files = [
    "Liga Argentina Fútbol.csv",
    "Messi_tweets.csv",
    "Partidos de clasificación desde 1960 hasta 2024.csv",
    "Ronaldo_tweets.csv",
]

# Convertir cada archivo CSV a JSON
for csv_file in csv_files:
    json_file = os.path.splitext(csv_file)[0] + ".json"  # Nombre del archivo JSON
    
    with open(csv_file, mode="r", encoding="utf-8") as file:
        reader = csv.DictReader(file)
        data = list(reader)  # Convertir a lista de diccionarios

    with open(json_file, mode="w", encoding="utf-8") as file:
        json.dump(data, file, indent=4, ensure_ascii=False)

    print(f"Convertido: {csv_file} → {json_file}")
```

```python
import json
import sqlite3
import os

# Lista de archivos JSON a importar
json_files = [
    "Liga Argentina Fútbol.json",
    "Messi_tweets.json",
    "Partidos de clasificación desde 1960 hasta 2024.json",
    "Ronaldo_tweets.json",
]

# Nombre de la base de datos
db_name = "futbol.db"

# Conectar a SQLite (crea el archivo si no existe)
conn = sqlite3.connect(db_name)
cursor = conn.cursor()

def limpiar_nombre(nombre):
    """Reemplaza espacios y caracteres especiales en nombres de tablas o columnas."""
    return nombre.replace(" ", "_").replace("-", "_").replace(".", "_")

for json_file in json_files:
    table_name = limpiar_nombre(os.path.splitext(json_file)[0])  # Limpiar nombre
    
    # Leer JSON
    try:
        with open(json_file, "r", encoding="utf-8") as file:
            data = json.load(file)
    except Exception as e:
        print(f"⚠️ Error al leer {json_file}: {e}")
        continue

    if not data:
        print(f"⚠️ Archivo vacío: {json_file}")
        continue
    
    # Obtener columnas desde la primera fila del JSON
    columns = list(data[0].keys())
    columns = [limpiar_nombre(col) for col in columns]  # Limpiar nombres de columnas
    columns_str = ", ".join(f'"{col}"' for col in columns)  # Encerrar en comillas dobles
    
    # Crear tabla en SQLite (si no existe)
    create_table_query = f"CREATE TABLE IF NOT EXISTS {table_name} ({', '.join([f'"{col}" TEXT' for col in columns])})"
    cursor.execute(create_table_query)

    # Insertar datos en la tabla
    for row in data:
        placeholders = ", ".join(["?"] * len(columns))
        values = tuple(row.get(col, "") for col in columns)  # Asegurar valores aunque falten claves
        insert_query = f'INSERT INTO {table_name} ({columns_str}) VALUES ({placeholders})'
        try:
            cursor.execute(insert_query, values)
        except sqlite3.Error as e:
            print(f"⚠️ Error al insertar en {table_name}: {e}")

    print(f"✅ Datos insertados en la tabla: {table_name}")

# Guardar cambios y cerrar conexión
conn.commit()
conn.close()
print(f"🎉 Base de datos '{db_name}' creada correctamente.")
```
