```python
from spellchecker import SpellChecker
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv("evento_musicales_MX.csv", encoding="ISO-8859-1")

# Inicializar el corrector ortográfico para español
spell = SpellChecker(language='es')

# Función para corregir palabras mal escritas, verificando que no sean None
def corregir_palabra(palabra):
    if not isinstance(palabra, str):  # Verificar si la palabra es una cadena de texto
        return palabra  # Si no es una cadena, devolverla tal cual (ej. NaN, None)
    
    # Dividir la palabra en partes si contiene varios errores
    palabras = palabra.split()
    palabras_corregidas = [spell.correction(palabra) or palabra for palabra in palabras]  # Evitar None
    return " ".join(palabras_corregidas)

# Aplicar la corrección ortográfica a la columna 'Nombre'
df['Nombre'] = df['Nombre'].apply(corregir_palabra)

# Verificar el cambio
print(df.head())
```
